```python
import pandas as pd

df = pd.read_csv('/content/indigo_customerfeedback.csv')
df.head()
```

#**Step 1: Prepare the Data**

```python
# Drop Duplicates
df1 = df.drop_duplicates(subset=['customer_id', 'route'], keep='first')
df1.head()
```

```python
#  all values in meal_type are in title case
df1['meal_type'] = df1['meal_type'].str.title()
df1.head()
```

```python
# all values in the route column to uppercase
df1['route'] = df1['route'].str.upper()
df1.head()
```

```python
""" 4. Fix values in `complaint_flag` column:
    - Standardize misspelled entries like `"yess"`  `"Yes"`, `"noo"`  `"No"`
    - Fill missing values with `"Not Available"
"""
df1['complaint_flag'] = df1['complaint_flag'].replace({'noo':'NO', 'yess':'Yes'}).str.capitalize().fillna('Not Available')
df1.head()
```

```python
# Fill missing values in taste_rating with a neutral default (use 3)
df1['taste_rating'] = df1['taste_rating'].fillna(3)
df1.head()
```

```python
df1.info()
```

#**Step 2: Analysis Report**

** Part A: Understanding the Problem Based on Taste Rating**

```python
low_taste_df = df1[df1['taste_rating'] < 3]
print("Number of passengers with taste rating < 3:", len(low_taste_df))
```

```python
provider_counts = low_taste_df['meal_provider'].value_counts()
print("Meal provider counts for taste rating < 3:")
provider_counts
```

```python
print("Rotes wtih feedbacks for taste rating < 3:")
low_taste_df[['feedback_id', 'route']]
```

```python
bad_feedback_df = df1[(df1['taste_rating'] <= 2) & (df1['freshness_rating'] <= 2)]
print("Feedback with taste and freshness ratings  2:")
bad_feedback_df
```

** Part B: Is the Problem Linked to a Specific Meal Provider?**

```python
complaints_df = df1[df1['complaint_flag'] == 'Yes']
complaints_per_provider = complaints_df['meal_provider'].value_counts()
print("Number of complaints per meal provider:")
complaints_per_provider
```

```python
passenger_low_taste = df1[(df1['taste_rating'] < 3) & (df1['meal_provider'] == 'LocalCaterer')]
print("Number of low taste ratings (<3) from LocalCaterer:", len(passenger_low_taste))
```

```python
low_packaging_rating = df1[(df1['packaging_rating'] <= 2) & (df1['meal_provider'] == 'SkyGourmet')]
print("Feedback with packaging rating  2 from SkyGourmet:")
print(low_packaging_rating['feedback_id'])
```

** Part C: Is the Problem Route-Specific?**

```python
low_taste_dellr = df1[(df1['taste_rating'] < 3) & (df1['route'] == 'DEL-BLR')]
print("Number of low taste ratings (<3) from DEL-BLR:", len(low_taste_dellr))
```

```python
complaints_bomhyd = df1[(df1['complaint_flag'] == 'Yes') & (df1['route'] == 'BOM-HYD')]
print("Number of complaints from BOM-HYD:", len(complaints_bomhyd))
```

```python
freshness_rating = df1[(df1['freshness_rating'] <= 2) & (df1['route'] == 'MAA-CCU')]
print("Feedback with freshness rating  2 from MAA-CCU:")
print(freshness_rating['feedback_id'])
```

** Part D: Evaluating the Packaging**

```python
low_packaging_df = df1[df1['packaging_rating'] < 3]
print("Number of meals with packaging rating < 3:", len(low_packaging_df))
```

```python
packaging_complaints = low_packaging_df[low_packaging_df['complaint_flag'] == 'Yes']
print("Number of flights that also had complaints raised wiht packaging < 3:", len(packaging_complaints))
```

```python
poor_provider = low_packaging_df['meal_provider'].value_counts().sort_values(ascending=False)
print("Meal providers with poor packaging ratings:")
poor_provider
```

** Part E: Are the Same Customers Complaining Often?**

```python
customer_count = df1['customer_id'].value_counts()
repeat_customer = customer_count[customer_count > 1]
print("Number of customers who have made more than one flight:", len(repeat_customer))
```

```python
repeat_df = df1[df1['customer_id'].isin(repeat_customer.index)]
poor_taste = repeat_df[repeat_df['taste_rating'] <= 2]
poor_taste_counts = poor_taste['customer_id'].value_counts()
multi_poor_taste_customers = poor_taste_counts[poor_taste_counts > 1]
print("Repeat customers who gave poor taste scores ( 2) more than once:", len(multi_poor_taste_customers))
```

#**Summary**

```python
# 1. Vendor with most complaints
top_complaint_vendor = df[df['complaint_flag'] == 'Yes']['meal_provider'].value_counts().idxmax()

# 2. Route with most low taste ratings
low_taste_routes = df[df['taste_rating'] < 3]['route'].value_counts()
top_low_taste_route = low_taste_routes.idxmax()



# Final Summary
print("## Final Insight:\n")
print(
    f"- Most complaints came from meals served by {top_complaint_vendor}, and route {top_low_taste_route} had the most low taste scores.\n"
    f"- Meals with poor taste and packaging ratings contributed significantly to complaints.\n"
    f"- {len(multi_poor_taste_customers)} customers had recurring poor experiences out of the 11 customers who gave feedback."
)
```
